```python
# Ejercicio 1

import numpy as np 
import matplotlib.pyplot as plt

#Punto 1

def qu_(phi,c,d,gamma,z): #phi angulo de fricciÃ³n, c cohesiÃ³n, d diÃ¡metro pilote, gamma peso efectivo, z profundidad punta 
  phi_ = np.radians(phi)
  k = np.radians(z/d)
  nq = np.exp(np.pi*np.tan(phi_))*(np.tan(np.pi/4+phi_/2)**2)
  nc = (nq-1)*(1/(np.tan(phi_)))
  sc = nq/nc + 1
  sq = 1 + np.tan(phi_)
  sigmav = gamma*z
  dc = 1 + (0.4*(np.tan(k)))
  dq = 1 + 2*np.tan(phi_)*((1-np.sin(phi_))**2)*(1/(np.tan(k)))
  qu = c*nc*sc*dc+sigmav*nq*sq*dq #tensiÃ³n Ãºltima en la punta
  qu_ = qu*(np.pi*d**2/4) #carga Ãºltima en la punta
  return qu_

qu_(1,4,3,6,7)

#Punto 2
_phi = np.linspace(15,35,40)
y = qu_(_phi,2,2,2.5,10) #c=2 d=2 gamma=2.5 z=10

plt.plot(_phi,y)
plt.xlabel("Ãngulo de fricciÃ³n")
plt.ylabel("Carga Ãºltima")
```
